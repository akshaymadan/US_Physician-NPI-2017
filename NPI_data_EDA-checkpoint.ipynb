**Segmenting by No of Beneficiaries**

```python
df.loc[(df['NumberofMedicareBeneficiaries'] > 0) & (df['NumberofMedicareBeneficiaries'] <= 50), 'Segments'] = 'Beneficiary Segment 5'
df.loc[(df['NumberofMedicareBeneficiaries'] > 50) & (df['NumberofMedicareBeneficiaries'] <= 100), 'Segments'] = 'Beneficiary Segment 4'
df.loc[(df['NumberofMedicareBeneficiaries'] > 100) & (df['NumberofMedicareBeneficiaries'] <= 200), 'Segments'] = 'Beneficiary Segment 3'
df.loc[(df['NumberofMedicareBeneficiaries'] > 200) & (df['NumberofMedicareBeneficiaries'] <= 400), 'Segments'] = 'Beneficiary Segment 2'
df.loc[(df['NumberofMedicareBeneficiaries'] > 400), 'Segments'] = 'Beneficiary Segment 1'
```

```python
pd.DataFrame(df['Segments'].value_counts())
```

```python
df['Payment_Segments'] = 'zzz'
```

```python
df['TotalMedicalMedicareStandardizedPaymentAmount']=df['TotalMedicalMedicareStandardizedPaymentAmount'].fillna(df['TotalMedicalMedicareStandardizedPaymentAmount'].median())
```

**Segmenting by Total Medical Medicare Standardized PaymentAmount**

```python
df.loc[(df['TotalMedicalMedicareStandardizedPaymentAmount'] >0) & (df['TotalMedicalMedicareStandardizedPaymentAmount'] <= 25000), 'Payment_Segments'] = 'Payment Segment 5'
df.loc[(df['TotalMedicalMedicareStandardizedPaymentAmount'] > 25000) & (df['TotalMedicalMedicareStandardizedPaymentAmount'] <= 50000), 'Payment_Segments'] = 'Payment Segment 4'
df.loc[(df['TotalMedicalMedicareStandardizedPaymentAmount'] > 50000) & (df['TotalMedicalMedicareStandardizedPaymentAmount'] <= 75000), 'Payment_Segments'] = 'Payment Segment 3'
df.loc[(df['TotalMedicalMedicareStandardizedPaymentAmount'] > 75000) & (df['TotalMedicalMedicareStandardizedPaymentAmount'] <= 100000), 'Payment_Segments'] = 'Payment Segment 2'
df.loc[(df['TotalMedicalMedicareStandardizedPaymentAmount'] > 100000) , 'Payment_Segments'] = 'Payment Segment 1'


```
